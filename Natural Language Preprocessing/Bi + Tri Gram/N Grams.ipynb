# Bi Gram

```python
def make_bi_unique(tokens):
    unique_bi_words = []

    for i in range(len(tweets)):
        temp = tweets[i].lower().split()
        for j in range(1, len(temp)):
            word = temp[j-1] + ' ' + temp[j]
            if word not in unique_bi_words:
                unique_bi_words.append(word)
    return unique_bi_words
```

```python
def bi_gram_func(tweets):
    vocab = make_bi_unique(tweets)
    bi_gram = []
    for i in range(len(tweets)):
        out =[]
        temp1 ={}
        temp = tweets[i].lower().split()
        for j in range(1, len(temp)):
            word = temp[j-1] + ' ' + temp[j]
            if word in temp1:
                temp1[word] += 1
            else: 
                temp1[word] = 1

        for j in vocab:
            if j in temp1:
                out.append(temp1[j])
            else:
                out.append(0)
        bi_gram.append(out)
    return pd.DataFrame(bi_gram, columns=vocab)
```
